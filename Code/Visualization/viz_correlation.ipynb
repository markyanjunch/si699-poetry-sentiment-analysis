## Correlation between VAD and 8 emotions

```python
import pandas as pd
import altair as alt
alt.data_transformers.disable_max_rows()
```

```python
df = pd.read_csv('scores.csv')
```

```python
df_VAD_EMO = df[['valence', 'arousal', 'dominance', 
                 'anger', 'fear', 'anticipation', 'trust', 'surprise', 'sadness', 'joy', 'disgust']]
df_VAD_EMO
```

```python
df_corr = df_VAD_EMO.corr().stack().reset_index().rename(columns={0: 'correlation', 'level_0': 'var1', 'level_1': 'var2'})
df_corr['corr_label'] = df_corr['correlation'].map('{:.3f}'.format)
df_corr
```

```python
df_corr = df_corr[df_corr['var1'].isin(['valence', 'arousal', 'dominance'])]
df_corr = df_corr[~df_corr['var2'].isin(['valence', 'arousal', 'dominance'])].reset_index(drop=True)
df_corr
```

```python
base = alt.Chart(df_corr).mark_rect().encode(
    x=alt.X(
        'var2:O', 
        title=None, 
        axis=alt.Axis(labelAngle=0, labelFontSize=14, labelFontWeight='bold', ticks=False, domainWidth=0)
    ),
    y=alt.Y(
        'var1:O', 
        title=None, 
        axis=alt.Axis(labelFontSize=14, labelFontWeight='bold', labelPadding=5, ticks=False, domainWidth=0)
    ),
    color=alt.Color(
        'correlation:Q', 
        title='correlation coefficient', 
        scale=alt.Scale(scheme='blues'),
        legend=alt.Legend(tickCount=2, orient='top', titleFontSize=12)
    )
).properties(
    width=800,
    height=300
)

text = base.mark_text(fontSize=13).encode(
    text='corr_label',
    color=alt.value('black')
)

(base+text).configure_view(strokeWidth=0)
```
