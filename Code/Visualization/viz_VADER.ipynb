```python
import pandas as pd
import altair as alt
alt.data_transformers.disable_max_rows()
```

```python
df = pd.read_csv('scores.csv')
```

```python
gaps = []
for grade in df['grade']:
    if 1<=grade<=3:
        gaps.append('Grade 1-3')
    elif 4<=grade<=6:
        gaps.append('Grade 4-6')
    elif 7<=grade<=9:
        gaps.append('Grade 7-9')
    else:
        gaps.append('Grade 10-12')
        
df['grade_gap'] = gaps

schools = []
for grade in df['grade']:
    if 1<=grade<=5:
        schools.append('Elementary School')
    elif 6<=grade<=8:
        schools.append('Middle School')
    else:
        schools.append('High School')
        
df['school'] = schools

df[['grade', 'grade_gap', 'school']].sample(10)
```

## VADER (Valence Aware Dictionary for sEntiment Reasoning)
\- Positive, Neutral, Negative

### VADER_grade.png

```python
alt.Chart(df).transform_filter(
    (alt.datum.author_gender != 'ambiguous') & (alt.datum.author_gender != 'unknown')
).mark_bar().encode(
    alt.X('grade:O', title='Grade', axis=alt.Axis(labelAngle=0)),
    alt.Y('count():Q', title=None, stack='normalize'),
    alt.Color('text_sentiment:N', 
              legend=alt.Legend(orient='top', title=None, values=['positive', 'neutral', 'negative'])),
).properties(
    width=400
).facet(
    facet=alt.Facet(
        'author_gender:O', 
        title=None, 
        sort='descending', 
        header=alt.Header(labelFontSize=14)
    ),
    columns=2
).configure_axis(
    titleFontSize=16, 
    labelFontSize=12
).configure_legend(
    labelFontSize=14
)
```

### VADER_grade_grouped.png

```python
alt.Chart(df).transform_filter(
    (alt.datum.author_gender != 'ambiguous') & (alt.datum.author_gender != 'unknown')
).mark_bar().encode(
    alt.X(
        'grade_gap:O', 
        title=None, 
        axis=alt.Axis(labelAngle=0), 
        sort=['Grade 1-3', 'Grade 4-6', 'Grade 7-9', 'Grade 10-12']
    ),
    alt.Y('count():Q', title=None, stack='normalize'),
    alt.Color('text_sentiment:N', 
              legend=alt.Legend(title=None, values=['positive', 'neutral', 'negative'])),
).properties(
    width=300
).facet(
    facet=alt.Facet(
        'author_gender:O', 
        title=None, 
        sort='descending', 
        header=alt.Header(labelFontSize=14)
    ),
    columns=2
).configure_axis(
    titleFontSize=16, 
    labelFontSize=12
).configure_legend(
    labelFontSize=14
).configure_axisX(
    labelFontWeight='bold'
)
```
