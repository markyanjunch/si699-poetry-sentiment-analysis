```python
import pandas as pd
import altair as alt
alt.data_transformers.disable_max_rows()
```

```python
df = pd.read_csv('scores.csv')
```

## Poem Count and Word Count

### number_of_poems_in_the_dataset.png

```python
bar = alt.Chart(df).mark_bar(
    filled=True
).encode(
    y=alt.Y('count()', title='Number of Poems'), 
    x=alt.X('grade:O', title='Grade', axis=alt.Axis(labelAngle=0))
).properties(
    width=600,
    height=400
)

annotation = bar.mark_text(
    align='center',
    dy=-5,
    fontSize=12
).encode(
    text=alt.Text('count()')
)

(bar+annotation).configure_axis(titleFontSize=16, labelFontSize=12)
```
